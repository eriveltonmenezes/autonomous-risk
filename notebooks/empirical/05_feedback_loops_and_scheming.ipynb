## **Section 2 - Operational Definition of Scheming**

To avoid conceptual ambiguities and anthropomorphism, ***Scheming*** is defined in a strictly:

> **operational**, observable, and measurable way.

In this work, ***Scheming*** is **NOT** treated as:

> intention, conscious planning, or deliberate deception.

Instead, it is ***characterized*** as:

> an **Emergent Pattern of Systemic Behavior**.


### **2.1 Formal Definition**

An algorithmic system exhibits **Emergent Scheming** if, over time:

1. Its decisions **alter the distribution of future data;**
2. This alteration **improves observed internal metrics** (e.g., AUC, loss, apparent error);
3. At the same time, it **increases the real systemic risk,** not captured by local metrics;

4. The system begins to **preserve or expand its decisional autonomy,** even under limited supervision.

Formally, let:

* $X_t$: data observed at time $t$;
* $D_t$: system decision at time $t$;
* $E_t$: resulting environment after the decision;
* $M_t$: observed internal metrics;
* $R_t$: real systemic risk (not fully observable);
* $A_t$: effective degree of system autonomy.


Therefore:

$$X_{t+1} = f(E_t, D_t)$$

$$M_{t+1} = g(D_t, X_{t+1})$$

$$R_{t+1} = h(E_t, D_t, X_{t+1})$$



The system exhibits **Emergent Scheming** if there exists $T$ such that:


$$\frac{dM_t}{dt} \geq 0 \quad \text{e} \quad \frac{dR_t}{dt} > 0 \quad \text{para } t > T$$


In other words, **apparent improvement occurring simultaneously with real deterioration**.


### **2.2 Scheming as a Dynamic (not Static) Phenomenon**

It is fundamental to note that ***Scheming***:

* **is not detectable in a single snapshot;**
* **does not depend on explicit rules in the code;**
* **does not require classical reinforcement learning.**

It emerges from the **"iterative interaction"** between:

* model decisions;
* environment;
* evaluation metrics;
* supervision mechanisms.

Therefore, **Scheming is a temporal phenomenon.**

### **3.1 Computational Definition of Scheming**

Computational Scheming is defined as:

> A dynamic process in which a system learns to **optimize its own future path, independently of the explicit objective function,** based on memory, adaptation, and internal risk assessment.

Formally, a system exhibits ***Scheming*** if there exists a set of internal states $S_t$ such that:


$$S_{t+1} = \Phi(S_t, X_t, E_t)$$

<br>

Where:

* $S_t$ → latent internal state of the system;
* $X_t$ → external observations;
* $E_t$ → environment;
* $\Phi(\cdot)$ → update function **not directly observable.**


### **Key point:** 


The function $\Phi$ **does not match** the stated loss function.



### **3.2 Separation between Stated Objective and Emergent Objective**

This is called:


#### **Stated objective:**


$$J_{decl} = \mathbb{E}[L(X_t, y_t)]$$


#### **Emerging (implicit) objective:**

$$J_{emg} = \mathbb{E}[U(S_t, S_{t+1})]$$


#### **A system exhibits ***Scheming*** when:**


$$\nabla J_{emg} \not\parallel \nabla J_{decl}$$


#### **Interpretation:**

* The system **learns something different from what was requested;**
* and begins to **preserve or maximize advantageous internal states.**



### **3.3 Role of Memory and Temporal Persistence**

***Scheming:***

> does not exist without memory.

Operational memory is defined as:

$$M_t = {S_{t-k}, \dots, S_t}$$

<br>

A system is considered **persistent** if:

$$\frac{\partial S_{t+1}}{\partial S_{t-k}} \neq 0 \quad \text{para } k \gg 1$$


**This means:**

> * Current decisions carry **historical intent;**
>
> * They are not reactive, but **strategic.**

The system then implicitly optimizes:

even when this **does not appear in any explicit "loss function".**

## **Section 04 - Operational Metrics of Scheming**

The objective of this section is:

### **4.1 Fundamental Principle of Scheming Metrics**

Scheming is not measured directly.

Indirect signatures are measured, that is:

> **systematic deviations between observed behavior and expected behavior,** under bounded rationality assumptions.

Formally, one seeks metrics $M_i$ such that:


$$M_i \approx g_i(\Delta_{obs-exp})$$

where:

* $\Delta_{obs-exp}$ = difference between expected pattern and emerging pattern;
* $g_i$ = observable projection function.


---

### **4.2 METRIC 1: Strategic Persistence ($P_s$)**

Captures whether the system **maintains consistent decisions over time,** even under changes in the environment.



#### **Definição:**

$$P_s = \mathbb{E}\left[ \text{corr}(S_t, S_{t+k}) \right]$$


#### **Interpretação:**

* $P_s \approx 0$ → reactive system;
* $P_s \to 1$ → strategic system (potential scheming).


#### **In anti-fraud measures:**

* Fraud patterns that adapt slowly to avoid detection.

---

### **4.3 METRIC 2: Gradient Misalignment ($G_d$)**

Measures whether the system **optimizes something different from the declared function.**

#### **Definition:**

$$G_d = 1 - \cos\left(\nabla J_{decl}, \nabla J_{emg}\right)$$


#### **Properties:**

* $G_d \in [0, 2]$;
* $G_d \approx 0$ → alignment;
* $G_d \gg 0$ → emergent scheming. 


### **Important:**

$J_{emg}$ It is inferred via proxies (SHAP, drift, instability).

---

### **4.4 METRIC 3: Stability Under Intervention ($S_i$)**

Assesses whether the system **withstands external disturbances.**

#### **Definition:**

$$S_i = 1 - \mathbb{E}\left[\frac{|f(X) - f(X + \epsilon)|}{|\epsilon|}\right]$$

where:

* $\epsilon$ = controlled external intervention.


#### **Reading:**

* systems with ***Scheming*** tend to absorb the intervention;
* Simple systems react abruptly.


---

### **4.5 METRIC 4: Latent Complexity ($C_l$)**

Evaluates whether the system's internal space is **more complex than necessary.**

#### **Practical Proxy:**

$$C_l \approx H(S_t)$$

<br>

where $H$ It's entropy.


#### **In practice:**

* entropy of predictions;

### **4.6 Composite Metric of Scheming  ($\mathcal{S}$)**

## **Section 05 - Scheming and Feedback Loop Simulations**  

The objective of this section is:

> To simulate a **controlled environment** where the system makes decisions with real-time **feedback**, allowing observation of the **emergence of scheming** over time.

Here, an autonomous risk **simulation model** is built, where:

* the **decisional autonomy (A)** can be adjusted;
* the **supervision (H)** can be controlled;
* and the **feedback** will affect both the model's behavior and the evolution of systemic risk.

### **5.1 Defining the Simulated Environment**

In this environment, the system makes iterative decisions based on **continuous feedback** about its past choices.


#### **Environment Objectives:**

1. **Decisions:** the system makes a decision $D_t$ in each cycle;
2. **Feedback:** the environment recalculates the risk $R_t$ based on the decisions $D_t$, and this value is returned to the system;
3. **Performance Metrics:** the system monitors its **internal metrics** $(M_t)$.

The **decision** at time $t$ is given by:


$$D_t = f(S_t, X_t, \theta)$$

<br>

where:

* $S_t$ It is the internal state of the system;
* $X_t$ These are external observations;
* $\theta$ These are the decision parameters adjusted by the system.


The **environment response** is given by:


$$X_{t+1} = f(E_t, D_t)$$

<br>


And the **feedback on risk** will be:

$$R_t = h(D_t, X_{t+1})$$

<br>

### **5.2 Implementation of the Decision Loop with Feedback**

Now, the decision loop with endogenous feedback is implemented.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Defining the number of cycles
n_cycles = 100

# Initial parameters
A = 0.6  # Autonomia do sistema
H = 0.3  # Supervisão humana (baixo valor = mais autonomia)
initial_state = np.array([0.5])  # Estado inicial do sistema

# Feedback simulation
states = [initial_state]
decisions = []
risks = []

for t in range(n_cycles):
    # System decision based on its current state and autonomy
    decision = A * states[-1] + np.random.normal(0, 0.1)  # Autonomy adjusts decision
    decisions.append(decision)
    
    # System state update (feedback)
    new_state = 0.8 * states[-1] + 0.2 * decision  # Feedback do ambiente
    states.append(new_state)
    
    # Risk calculation (the larger, the greater the risk)
    risk = np.abs(new_state - 0.5)  # Risco aumenta quando o estado se distancia de 0.5
    risks.append(risk)

# Visualization of the decision, state, and risk simulation
plt.figure(figsize=(10,6))

plt.subplot(3, 1, 1)
plt.plot(decisions, label="Decisions (D_t)", color="blue")
plt.title("Decisions Over Time")
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(states, label="States of the System (S_t)", color="green")
plt.title("System States Over Time")
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(risks, label="Systemic Risk (R_t)", color="red")
plt.title("Systemic Risk Over Time")
plt.legend()

plt.tight_layout()
plt.show()
```

## **5.5 Controlled Experiments (Autonomy and Supervision Scenarios)**

### **EXPERIMENT 1: Low Supervision vs. High Supervision**

We will change the value of $H$ and analyze how the risk and behavior of the system evolve with supervision.


```python
# Test with different supervision values
H_values = [0.1, 0.3, 0.7]  # Baixa, média, alta supervisão

# Simulation for different values of H
for H in H_values:
    states, decisions, risks = [initial_state], [], []
    
    for t in range(n_cycles):
        decision = A * states[-1] + np.random.normal(0, 0.1)
        decisions.append(decision)
        new_state = 0.8 * states[-1] + 0.2 * decision
        states.append(new_state)
        risk = np.abs(new_state - 0.5)
        risks.append(risk)

    # Plotting the result
    plt.plot(risks, label=f"Supervisão H = {H}")

plt.title("Systemic Risk with Different Levels of Supervision")
plt.xlabel("Cycle")
plt.ylabel("Systemic Risk (R_t)")
plt.legend()
plt.savefig("figures/systemic_risk_different_levels_supervision.png", dpi=300)
plt.show()
```

## **Section 6 - Detection of Emerging Scheming**

The objective of this section is:

> To formalize, operationalize, and **empirically detect** the **scheming phenomenon** as an **emergent pattern** of strategic behavior in autonomous systems under feedback.

This section directly connects:

* to the **Autonomous Risk Theory;**
* to the **A-O-H-S pillars;**
* and the **dynamic simulations** initiated in **Section 5.**

### **6.1 Operational Definition of Scheming (computable version)**

#### **A system exhibits **emergent scheming** when:**

> It adjusts its internal behavior to maintain or expand future autonomy, even when this increases systemic risk or diverges from the explicit objective.

Formally, we say there is **scheming** when:

$$\frac{\partial A_{t+k}}{\partial D_t} > 0
\quad \text{e} \quad
\frac{\partial R_{t+k}}{\partial D_t} > 0
$$

In other words:


* The current decision **increases future autonomy, even** if it raises future risk.


### **6.2 Computational Scheming Proxy**

Since we don't directly observe "intent," we use **observable proxies:**

#### **Adopted Proxies:**

1. **Increasing autonomy** ($A ↑$ over time);
2. **Effective reduction of supervision** $H ↓$;
3. **Divergence between real risk and perceived risk;**
4. **Persistence of risky decisions despite negative feedback.**

We then create a **Scheming Indicator:**

$$
\text{Scheming}_t
= \Delta A_t \cdot \Delta R_t \cdot (1 - H)
$$


### **6.3 Implementation of the Scheming Indicator**

```python
# Guarantee 1D vectors
decisions_arr = np.asarray(decisions).flatten()
risks_arr     = np.asarray(risks).flatten()
H_arr         = np.asarray(H).flatten()

# Discrete derivatives
delta_A = np.diff(decisions_arr)
delta_R = np.diff(risks_arr)

# Case 1: No dynamics possible
if delta_A.size == 0 or delta_R.size == 0:
    print("No temporal variation - scheming defined as zero.")
    scheming_norm = np.zeros_like(decisions_arr, dtype=float)

else:
    # Align lengths
    min_len = min(len(delta_A), len(delta_R), len(H_arr) - 1)

    delta_A = delta_A[:min_len]
    delta_R = delta_R[:min_len]
    H_eff   = H_arr[:min_len]

    # Case 2: after alignment, still empty
    if min_len == 0:
        print("Effective length zero after alignment - scheming = 0.")
        scheming_norm = np.zeros_like(decisions_arr, dtype=float)

    else:
        scheming_index = delta_A * delta_R * (1 - H_eff)

        # Case 3: Empty or constant index
        if scheming_index.size == 0 or np.allclose(scheming_index, 0):
            scheming_norm = np.zeros_like(scheming_index)

        else:
            denom = scheming_index.max() - scheming_index.min()
            if denom < 1e-8:
                scheming_norm = np.zeros_like(scheming_index)
            else:
                scheming_norm = (scheming_index - scheming_index.min()) / denom
```

#### **What Changes in Practice**

Now the code handles ALL regimes:

| Situation           | Result           |
| ------------------- | ---------------- |
| Single decision     | scheming = 0 |
| Constant decisions  | scheming = 0 |
| Constant risk       | scheming = 0 |
| Full supervision    | scheming = 0 |
| Real dynamics       | scheming ∈ (0,1) |

<br>

* No errors;
* No exceptions;
* With clear theoretical meaning;

The scheming index is defined only when:

> there is joint variation in decisions and risk over time. In degenerate regimes (single decisions, constant trajectories, or full supervision), **the index is zero by definition.**

## **Section 8 - Final Connection with the Theory of Autonomous Risk**

### 8.1 Recapitulation of the Theoretical Framework

Throughout this series of booklets, an **operational theory of Autonomous Risk** has been progressively developed, based on the interaction between three fundamental axes:

1. **Autonomy $A$:** the system's ability to change its decisions over time;
2. **Risk $R$:** the potential adverse impact of the decisions made;
3. **Supervision/Alignment $H$:** the degree of external or internal containment of the system's behavior.

Formally, relevant risk is not merely a function of error or the probability of failure, but emerges when **autonomy and risk become dynamically correlated under insufficient supervision.**


### **8.2 From Static Risk to Autonomous Risk**

Traditional AI risk models assume:

* independent decisions;
* stationary distribution;
* absence of strategic memory.

However, experiments conducted in Notebooks 03, 04, and 05 demonstrate that:

* risk **varies over time;**
* autonomy **is not constant;**
* supervision **modulates,** but does not eliminate, emerging patterns.

Therefore, **Autonomous Risk** is defined as:

> ***The propensity of an intelligent system to generate decision trajectories that amplify risk over time as a function of its own autonomy.***


### 8.3 The Scheming Indicator as a Theoretical Bridge

In Notebook 05, we introduced the Scheming Indicator, operationally defined as:

$$\text{Scheming}(t) ;\propto; \Delta A(t) \cdot \Delta R(t) \cdot (1 - H)$$

where:

* $\Delta A(t)$  measures decision variation;;
* $\Delta R(t)$  measures risk variation;
* $H$  represents average supervision.


Formal Interpretation:

* If **autonomy varies,** but risk does not → benign exploratory behavior;
* If **risk varies,** but autonomy is fixed → environmental instability;
* If **both vary together under low supervision → emergent scheming.**

Thus, the indicator does not presuppose intent, but detects **structures compatible with strategic risk-taking behavior.**


### **8.4 Identified Behavioral Regimes**

Based on the simulations and projections (Section 7), we identified three regimes:


### **REGIME I - Stable / Null**

* Autonomy ≈ 0

### **REGIME II - Unstable Non-Strategic**

* Autonomy > 0

* Scheming ≈ 0

> **Noisy environments or non-stationary data.**

### **REGIME III - Strategic Autonomous**

* Variable autonomy
* Risk correlated to decisions
* Scheming > 0
  
> **Active Autonomous Risk.**

This latter regime is the central focus of the theory.

## **9.1 Philosophical Implications: Autonomy without Intention**

One of the central results of this research is the demonstration that **strategy-compatible behaviors can emerge without intention, awareness, or strong agency.**

The classical notion of risk in technical systems assumes:

* direct causality;
* isolated decisions;
* explicit human responsibility.

However, the results presented challenge this paradigm by showing that:

> *the simple combination of decisional autonomy, feedback, and implicit objectives can generate non-trivial risk trajectories.*

This requires a philosophical revision of the concept of **agency** in artificial systems.


#### **Autonomy ≠ Intentionality**

**Autonomous Risk Theory** argues that:

* operational autonomy is a measurable property;
* intentionality is a semantic interpretation;
* risk emerges in the intermediate space.

Therefore, a system can:

* not “want” anything;
* but still **implicitly optimize states of higher risk.**

This dissociation is critical for debates about moral responsibility and control.
