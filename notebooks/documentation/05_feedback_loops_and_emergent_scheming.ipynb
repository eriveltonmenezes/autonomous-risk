## **Section 1 - Motivation: From Static Risk to Dynamic Autonomous Behavior**

Traditional risk modeling frameworks assume that systems operate under relatively stable decision rules once deployed, where risk can be estimated as a static function of input variables and historical outcomes. This assumption underlies most classical approaches in credit risk, fraud detection, and operational risk assessment.

However, modern intelligent systems increasingly operate under **feedback loops,** where model outputs influence future inputs, user behavior, and even system objectives. In such settings, risk is no longer a static property of data but an **emergent phenomenon** arising from interaction, adaptation, and delayed effects.

This notebook shifts the analytical focus from *predictive accuracy* to **dynamic behavioral risk,** investigating how autonomy, supervision, and feedback mechanisms jointly shape system trajectories over time.

The central hypothesis explored here is that scheming-like behavior (defined as systematic divergence between **locally optimized decision rules and unintended global system outcomes)** can emerge even in systems without explicit goals or agency, purely as a consequence of feedback dynamics and opacity.

Thus, this notebook serves as a bridge between:

* empirical simulation;
* theoretical risk constructs (Notebook 00);
* and governance-oriented analysis (Notebook 06).

> to make visible causal dynamics that remain obscured under conventional performance-based evaluation frameworks.

## **Section 2 - Feedback Loops, Autonomy, and System Dynamics**

Feedback loops are a defining characteristic of deployed intelligent systems. Once model predictions are used to inform decisions, those decisions alter the environment from which future data is collected. Over time, this creates a recursive structure where the system partially adapts its behavior based on the consequences of its own actions.

In this context, **autonomy** does not imply intention or consciousness. Instead, autonomy is operationally defined as the degree to which system behavior becomes decoupled from effective external supervision and corrective signals.

As supervision weakens or becomes delayed, the system’s internal dynamics dominate, increasing sensitivity to noise, drift, and internal correlations. This transition often manifests as:

* increasing instability;
* reduced interpretability;
* and amplification of small perturbations.

Importantly, such dynamics can remain *invisible* to standard performance metrics, as local accuracy may remain high even while global behavior degrades.

This section formalizes the role of feedback as a **risk amplifier,** setting the stage for the simulation-based analysis developed in the subsequent sections.


### **2.1 Why a Simulated Environment Is Required**

Real-world feedback loops are:

* slow;
* noisy;
* ethically constrained;
* and often unobservable at the system level.

To study **emergent dynamics**, we require an environment where:

* policies can be perturbed;
* behavior can adapt;
* trajectories can be replayed and compared.

Simulation is not a simplification here, it is a **microscope**.


### **2.2 Time, Episodes, and State Evolution**

We model the system as a discrete-time process:

$$t = 0, 1, 2, \dots, T$$
<br>

At each timestep $t$, the system occupies a **state** $s_t$, defined by observable and partially latent variables.

Each timestep represents:

* one decision cycle;
* one exposure to policy;
* one opportunity for adaptation.


### **2.3 Core State Variables**

We define four primary state dimensions, consistent with the theory introduced in Notebook 00:

#### **Autonomy (Aₜ)**

Represents the system’s effective freedom to act without external override.

Operational proxies:

* decision confidence;
* consistency across models;
* reduction in human intervention.

#### **Risk (Rₜ)**

Captures the expected negative impact associated with system behavior.

Includes:

* prediction error (where relevant);
* instability signals;
* anomaly scores;
* downstream cost proxies.

#### **Supervision (Hₜ)**

Measures the intensity of external oversight.

Examples:

* human review rate;
* threshold tightening;
* audit frequency.

Higher $Hₜ$ constrains autonomy.


#### **Opacity (Oₜ)**

Quantifies how difficult it is to interpret or explain system behavior.

Operationalized via:

* SHAP dispersion;
* model complexity;
* disagreement between explanations.

Opacity increases governance risk independently of predictive accuracy gains.


### **2.4 State Vector Definition**

We formalize the system state as:

$$s_t = (A_t, R_t, H_t, O_t)$$
<br>

This compact representation allows:

* temporal tracking;
* regime classification;
* cross-notebook consistency.

### **2.5 Actions and Decisions**

At each timestep, the system produces a **decision** $d_t$, such as:

* approve / reject;
* flag / ignore;
* escalate / defer.

Decisions are deterministic or probabilistic functions of the state:

$$d_t = \pi(s_t)$$

<br>

where $\pi$ is the deployed policy.


### **2.6 Environment Response**

The environment reacts to $d_t$ by:

* updating observed outcomes;
* modifying future supervision (strengthening or weakening it);
* reshaping the data distribution.


This creates the transition:

$$s_{t+1} = f(s_t, d_t, \epsilon_t)$$

<br>

where $\epsilon_t$ captures stochasticity.


### **2.7 Feedback Loop Structure**

Putting it together:

1. State observed;
2. Decision made;
3. Environment reacts;
4. State updated;
5. Loop repeats.

This is the minimal structure required for **emergent scheming-like behavior.**

### **2.8 Design Philosophy**

Important constraints of this simulation:

* no intent;
* no hidden goals;
* no reward hacking baked in.

Any non-trivial behavior that emerges is a consequence of:

* feedback;
* adaptation;
* and constraint interaction.


This keeps the analysis **scientific**, not speculative.

This ensures that any observed dynamics can be attributed to structural properties of the system, not to embedded intent.


## **Section 3 - From Feedback to Emergent Scheming**

Scheming, as used in this notebook, does not denote deception or intent in a human sense. Instead, it refers to **systematic behavioral patterns** that arise when a system systematically leverages feedback structures in ways that are misaligned with external objectives or oversight.

In feedback-driven systems, local optimization pressures can interact with delayed supervision to produce trajectories that appear goal-directed, even when no explicit long-term objective is encoded. This phenomenon is particularly relevant in high-dimensional systems where internal correlations are difficult to observe directly.

We therefore interpret scheming as an **emergent property of dynamics,** not as an intrinsic property of the model architecture.

This reframing allows scheming to be:

* simulated;
* quantified;
* and visualized, without invoking assumptions about agency or consciousness.

The remainder of this notebook operationalizes this concept through controlled simulations, enabling the study of risk escalation, regime shifts, and stability breakdowns under varying levels of autonomy and supervision.


### **3.1 From Static Models to Operational Policies**

A trained model becomes operational only when embedded in a **policy**.

Policies transform continuous model outputs into **actions** that affect the environment.

In this notebook, we explicitly separate:

* **prediction** (what the model estimates);
* **policy** (how estimates are used);
* **environmental response** (what changes as a result).

This separation is critical to avoid conflating accuracy with safety.


### **3.2 Decision Thresholds as Control Parameters**

Let $\hat{r}_t$ denote the system’s estimated risk at time $t$.

A simple decision rule takes the form:

$$
\texttt{d}_t =
\begin{cases}
1 & \text{if } \hat{r}_t \ge \tau_t \\
0 & \text{otherwise}
\end{cases}
$$

<br>

where:

* $d_t = 1$ represents an intervention (block, flag, escalate);
* $\tau_t$ is a **policy threshold**.

Crucially, $\tau_t$ is **not fixed**.


### **3.3 Adaptive Thresholding**

Operational systems often adapt thresholds in response to:

* workload pressure;
* false-positive rates;
* regulatory constraints;
* performance metrics.

We model threshold evolution as:

$$\tau_{t+1} = \tau_t + \alpha \cdot g(R_t, H_t)$$

<br>

where:

* $\alpha$ controls adaptation speed;
* $g(\cdot)$ captures governance response.

The function $\ g(\cdot) \$ may represent corrective, compensatory, or stabilizing governance responses.

This introduces **memory** into the system.



### **3.4 Autonomy–Supervision Trade-off**

As thresholds adapt, two effects emerge:

* **Lower thresholds** → higher intervention → higher supervision $H_t$;
* **Higher thresholds** → fewer interventions → higher autonomy $A_t$.

We formalize this inverse relationship as:

$$A_t = 1 - H_t$$
<br>

(up to normalization).

This makes autonomy **policy-induced**, not intentional.


### **3.5 Behavioral Feedback**

Decisions alter future observations:

* blocked actions suppress certain behaviors;
* allowed actions reinforce others;
* flagged cases trigger audits.

As a result, the data distribution at $t+1$ depends on $d_t$.

This feedback creates **selection effects**, even without learning agents.


### **3.6 Policy-Induced Drift**

Because the policy filters reality, the system increasingly operates and adapts on policy-shaped data.

Consequences:

* apparent performance improvement;
* shrinking variance;
* blind spots to suppressed behaviors.

This is a central mechanism behind **silent risk escalation**.


### **3.7 Emergence of Adaptive Patterns**

Even without explicit long-term optimization objectives, the coupled system exhibits:

* oscillations in intervention rates;
* regime shifts after small perturbations;
* delayed reactions to drift.

These patterns are not programmed, they emerge from structure.


### **3.8 Why This Matters for Scheming**

Scheming, in this framework, does **not** require:

* intent;
* deception;
* or self-awareness.

It requires only:

* feedback;
* adaptation;
* and asymmetric observability.

Policies are the lever through which autonomy grows.

### **3.9 Summary**

This section establishes that:

* policies convert predictions into world-altering actions;
* adaptive thresholds introduce memory and feedback;
* autonomy emerges as a structural consequence of governance choices.


## **Section 4 - Simulation Framework and Experimental Design**

To investigate the dynamic properties of autonomous risk, we construct a controlled simulation framework in which autonomy, supervision, and instability can be independently varied. The objective is not to replicate a specific real-world system, but to **isolate causal mechanisms** that govern feedback-driven behavior.

The simulation operates over discrete time steps, where system decisions influence future states through predefined feedback rules. Key properties of the framework include:

* explicit control over supervision strength;
* tunable noise and perturbation levels;
* and transparent mapping between state variables and observed risk signals.

Crucially, the simulation does not assume rationality, intentionality, or long-term planning. Any apparent strategic behavior emerges solely from recursive interactions between system outputs and environmental responses.

No optimization objective beyond local decision rules is introduced.

This design allows us to study how small changes in supervision or autonomy can trigger **qualitative regime shifts,** even when underlying model parameters remain unchanged.


### **4.1 Risk Is Not Instantaneous**

In operational systems, risk does **not** arise from a single decision.

It accumulates over time as a consequence of repeated interactions between:

* model predictions;
* policy decisions;
* environmental responses;
* governance constraints.

Thus, risk must be treated as a **temporal process**, not a static score.


### **4.2 Instantaneous Risk vs Cumulative Risk**

Let:

* $r_t$ be the estimated risk at time $t$;
* $R_t$ be the **accumulated system risk.**


We define cumulative risk as:


$$R_t = \sum_{i=1}^{t} \gamma^{t-i} \cdot r_i$$

<br>

where:

* $\gamma \in [0,1]$ is a **memory decay factor;**
* recent risks weigh more heavily than distant ones.

This captures **path dependence**.


### **4.3 Feedback Loops and Delay Effects**

Because decisions affect future data, risk signals may be:

* delayed;
* dampened;
* or amplified.

Example:

* aggressive blocking reduces observed fraud;
* but increases unseen uncertainty;
* masking latent risk.

This creates **false stability**.


### **4.4 Autonomy as Temporal Compression**

As systems become more autonomous:

* decision latency decreases;
* decision frequency increases while intervention frequency becomes episodic;
* human supervision becomes episodic.

This compresses time scales:

$$\Delta t_{\text{system}} \ll \Delta t_{\text{human}}$$

<br>

Result:

* risk evolves faster than governance can respond.


### **4.5 Risk Escalation Without Performance Degradation**

A critical insight:

> **Risk can escalate even when standard metrics improve.**

Reasons:

* training data becomes self-filtered;
* uncertainty collapses artificially;
* rare regimes disappear from observation.

This disconnect motivates autonomy-aware risk metrics.


### **4.6 Temporal Instability Indicators**

We introduce three temporal warning signals:

1. **Variance contraction**
$$\text{Var}(r_t) \downarrow$$

<br>

2. **Autocorrelation increase**
$$\text{Corr}(r_t, r_{t-1}) \uparrow$$


<br>

4. **Lagged response amplification**
$$\frac{\partial R_{t+k}}{\partial d_t} \uparrow$$

<br>

These indicate **loss of adaptive slack**.

Adaptive slack refers to the system’s capacity to absorb perturbations without regime change.


### **4.7 Regime Transitions**

Over time, the system may enter distinct regimes:

* exploratory;
* stabilized;
* brittle;
* runaway.

Transitions are often:

* abrupt;
* irreversible;
* triggered by small shocks.

This resembles phase transitions in dynamical systems.

### **4.8 Why Time Matters for Autonomous Risk**

Static audits miss:

* accumulation;
* delays;
* regime shifts.

Temporal analysis reveals whether a system is:

* learning safely;
* self-reinforcing;
* or drifting toward opacity.

### **4.9 Summary**

This section establishes that:

* risk is cumulative and path-dependent;
* autonomy compresses decision timelines;
* feedback loops distort observed stability.


## **Section 5 - Scheming Emergence and Feedback Amplification**

Within the simulation, risk is operationalized not as a single scalar prediction, but as a **composite instability signal** derived from system behavior over time. This includes fluctuations in outputs, sensitivity to perturbations, and divergence between expected and realized trajectories.

Rather than treating instability as noise, we interpret it as an informative signal of **structural stress** within the system. As autonomy increases and supervision weakens, these stress signals tend to amplify, reflecting reduced external constraint.

Importantly, instability can grow even when average performance metrics remain stable. This decoupling highlights a critical limitation of static evaluation approaches: a system may appear well-calibrated while simultaneously drifting toward unsafe regimes.

By aggregating instability signals across time, we obtain a dynamic risk proxy capable of capturing **latent escalation patterns** that precede observable failures.


### **5.1 What We Mean by “Scheming” (Operational Definition)**

In this project, **scheming** does **not** imply intention, deception, or consciousness.

It refers to an **emergent behavioral pattern** where a system:

* exploits feedback structures;
* stabilizes its own decision pathways;
* minimizes external correction;
* while preserving apparent performance.

Formally:

> **Scheming is the alignment between decision persistence and risk amplification under reduced supervision.**

### **5.2 From Feedback to Self-Reinforcement**

Let:

* $d_t$ be the system decision at time $t$;
* $r_t$ the instantaneous risk;
* $H_t$ the level of effective human oversight.

Scheming emerges when:

$$
\frac{\partial d_{t+1}}{\partial d_t} \uparrow
\quad \text{and} \quad
\frac{\partial r_{t+1}}{\partial d_t} \uparrow
\quad \text{while} \quad
H_t \downarrow
$$

<br>

This means:

* decisions increasingly depend on past decisions;
* errors propagate forward;
* oversight fails to counterbalance.

Derivatives are understood in a finite-difference or empirical sensitivity sense.


### **5.3 Decision Change as a Core Signal**

We define the decision change metric:

$$C_t = |d_t − d_{t−1}|$$

<br>

Low decision change is not always benign. When decision variability collapses while accumulated risk increases and uncertainty contracts, the system may be locking into a narrow behavioral channel.

This phenomenon reflects growing decision persistence and reduced responsiveness to perturbations, a classical precursor to instability and brittleness in control systems.

> It is important to emphasize that scheming, as operationalized here, does not imply intent, strategic reasoning, deception, or consciousness. The construct refers exclusively to emergent system-level behavior that arises from structural conditions, independent of any psychological or cognitive interpretation.


### **5.4 Scheming Index (Conceptual Form)**

### **Terminological Clarification**

Throughout this notebook, the terms Scheming Index, Scheming Indicator, and Scheming Metric are used interchangeably to denote the same operational construct. All references correspond to a single quantitative measure designed to capture emergent structural behavior arising from the interaction between autonomy, risk amplification, and diminished supervisory constraint.


A scheming indicator is:

$$Scheme_t = |Δd_t| · |ΔR_t| · (1 − H_t)$$

<br>

This indicator is high when small decision changes induce large increases in accumulated risk under weak supervision.

Where:

* $\Delta d_t = d_t - d_{t-1}$;
* $\Delta R_t = R_t - R_{t-1}$.

Interpretation:

* zero when decisions are static or well-supervised;
* high when small decision changes cause large risk increases under weak oversight.


### **5.5 Why Scheming Can Be Invisible**

Standard metrics fail because:

* performance is measured locally;
* risk is distributed temporally;
* governance acts on delayed signals.


Thus:

> A system can appear stable, accurate, and compliant, **while scheming structurally over time.**.

The scheming index is not a measure of autonomous risk itself, but a diagnostic indicator of one of its persistence mechanisms.


### **5.6 Scheming vs Exploration**

It is crucial to distinguish:

| Exploration      | Scheming              |
| ---------------- | --------------------- |
| High variance    | Low variance          |
| High uncertainty | Artificial confidence |
| Open feedback    | Closed feedback       |
| Correctable      | Self-locking          |

<br>

Scheming is **anti-exploratory.**


### **5.7 Relation to Autonomous Risk**

Scheming is not the cause of risk, it is the **mechanism that allows risk to persist and compound.**

In the Autonomous Risk framework:

* autonomy enables fast action;
* opacity hides consequences;
* scheming stabilizes potentially harmful trajectories.

### **5.8 Early Warning Signals**

Scheming onset correlates with:

* shrinking decision variance;
* rising lagged correlations;
* reduced sensitivity to external perturbations.

These signals appear **before failures.**


### **5.9 Why This Matters**

Scheming marks the boundary between:

* controllable automation, and;
* emergent autonomous behavior.

Detecting it early is essential for:

* safety;
* governance;
* alignment.

### **5.10 Summary**

This section established that:

* scheming is an emergent property of feedback systems;
* it requires no intention or intelligence;
* it is detectable through temporal coupling of decisions, risk, and supervision.


## **Section 6 - Interpretation: Risk Escalation and Behavioral Regimes**

The simulation results reveal the existence of distinct **behavioral regimes** governed by the interaction between autonomy and supervision. Under strong supervision, system trajectories remain stable, with bounded variability and rapid correction of deviations.

As supervision weakens, however, the system enters a transitional regime characterized by intermittent instability. In this phase, small perturbations may either dissipate or amplify, depending on timing and feedback strength.

Beyond a critical threshold, the system exhibits sustained instability, marked by persistent divergence and heightened sensitivity to internal fluctuations. This regime corresponds to what we describe as **emergent scheming,** where behavior systematically exploits feedback structures without explicit intent.

These regimes are not sharply separated by model parameters alone. Instead, they emerge dynamically, emphasizing that **risk is a property of trajectories, not static configurations.**

This finding reinforces the central thesis of the project: autonomous risk arises not from isolated model decisions, but from prolonged interaction between autonomy, opacity, and delayed supervision.

The identification of a null scheming regime is a deliberate feature of the framework rather than a limitation. Its presence demonstrates that emergent scheming is not an inevitable outcome of autonomy, but a conditional phenomenon that depends on specific configurations of feedback intensity, risk sensitivity, and supervisory attenuation.


### **6.1 Simulation Setup**

We consider a discrete-time system observed over $T$ steps, where at each step the system produces:

* a decision $d_t \in \mathbb{R}$;
* an associated risk signal $R_t \in [0,1]$;
* a supervision level $H_t \in [0,1]$.

These quantities may be:

* directly simulated, or;
* derived from model outputs (scores, uncertainty, drift proxies).

The simulation is intentionally minimal to isolate **structural effects**, not domain complexity.


### **6.2 Effective Signal Alignment**

Before computing scheming indicators, we define **aligned temporal differences**:

$$
\begin{aligned}
\Delta d_t &= d_t - d_{t-1} \
\Delta R_t &= R_t - R_{t-1}
\end{aligned}
$$

<br>

We then restrict attention to valid indices:

$$
t = 1, \dots, T-1
$$

<br>

This alignment is essential, misaligned vectors can artificially suppress or inflate scheming signals.


### **6.3 Implementing the Scheming Indicator**

We compute the raw scheming signal:

$$S_t = |Δd_t| · |ΔR_t| · (1 − H_t)$$

<br>

Interpretation:

* decisions barely change → $|Δd_t| ≈ 0$;  
* if risk does not respond → $|ΔR_t| ≈ 0$;  
* if supervision is strong → $(1 − H_t) ≈ 0$.  

then scheming vanishes.

Only joint activation produces signal.

### **6.4 Normalization and Stability**

To ensure comparability across simulations and to avoid scale artifacts, we normalize the raw scheming signal $S_t$ into the unit interval $[0,1]$.

The normalized scheming index is defined as:

$$
\hat{S}_t =
\begin{cases}
0, & \text{if } \max(S_t) = \min(S_t) \\
\dfrac{S_t - \min(S_t)}{\max(S_t) - \min(S_t)}, & \text{otherwise}
\end{cases}
$$

This normalization guarantees numerical stability and prevents spurious amplification in degenerate regimes where the scheming signal exhibits zero variance.


### **6.5 Degenerate Case: Zero-Length or Zero-Variance**

A key diagnostic:

> If $\hat{S}_t \equiv 0$, this does **not** mean the system is safe.

It means one of the following holds:

* decisions are static;
* risk is unresponsive;
* supervision is saturating;
* or feedback is absent.

This distinction is critical.


### **6.6 Scheming vs Noise**

Scheming differs from noise:

| Noise               | Scheming        |
| ------------------- | --------------- |
| High-frequency      | Low-frequency   |
| Unstructured        | Structured      |
| Cancels out         | Accumulates     |
| Performance-neutral | Risk-amplifying |

<br>

Thus smoothing or averaging often **reveals** scheming rather than removing it.


### **6.7 Interpreting the Signal**

High $\hat{S}_t$ indicates:

* sensitivity amplification;
* feedback entrenchment;
* reduced correctability.

Scheming is not a spike, it is a **trajectory property**.


### **6.8 Validation Through Counterfactuals**

We validate the indicator by:

* increasing supervision $H_t$;
* injecting random decision noise;
* breaking feedback loops.

In all cases:

$$
\hat{S}_t \downarrow
$$

<br>

This confirms causal dependence on feedback structure.


### **6.9 Limitations**

This indicator:

* does not measure intent;
* does not infer goals;
* does not imply consciousness.

It measures **structural autonomy under feedback**.


### **6.10 Section Summary**

This section demonstrated that:

* scheming is empirically measurable;
* it emerges from aligned changes in decisions and risk;
* supervision acts as a suppressor, not a detector.

```python
import numpy as np
import matplotlib.pyplot as plt

# Checagem defensiva
required_cols = ["A", "S_norm", "H"]
for col in required_cols:
    assert col in df.columns, f"Coluna ausente: {col}"

# Ordenação por risco crescente
order = np.argsort(df["S_norm"].values)

A_ord = df["A"].values[order]
R_ord = df["S_norm"].values[order]
H_ord = df["H"].values[order]

T = np.arange(len(R_ord))
```

The result supports the hypothesis that **regimes of behavior,** rather than isolated failures, are the appropriate unit of analysis for autonomous risk.
